## Ejercicio 09

Cree un arreglo, de 8 columnas y  8 filas, ambas seleccionadas aleatoriamente (utilice el número de su documento de identidad como semilla), a partir del arreglo_00.

Luego cree un arreglo de dimensión 1x8 con número reales aleatorios en el intervalo 0,10 (utilice el número de su documento de identidad como semilla).

Asuma que el primer arreglo representa los coeficientes de un sistema de ecuaciones lineales, y el segundo arreglo contiene las respectivas constantes. Encuentre la solución al sistema utilizando funciones de álgebra lineal de NumPy.

Muestre el valor de cada una de las variables que resuelve el sistema de ecuaciones.

```python
# Importar librerías necesarias
# Librerías estándar de Python
from IPython.display import display, Latex

# Librerías de terceros
import numpy as np
from colorama import Fore, Style

# Crear un arreglo de 8 columnas y 8 filas seleccionadas aleatoriamente
np.random.seed(1001140290)
arreglo_12 = arreglo_00[np.random.randint(0, 1000, 8), :8].copy()

# Crear un arreglo de 1x8 con números reales aleatorios en el intervalo 0, 10
arreglo_13 = np.random.uniform(0, 10, size=8)

# Sistema de ecuaciones en LaTeX
print(Fore.CYAN + Style.BRIGHT + 
      "Sistema de ecuaciones:" + Style.RESET_ALL)
# Generar LaTeX para el sistema de ecuaciones
latex_str = r"\begin{align*}"

for i, row in enumerate(arreglo_12):
    equation = " + ".join([f"{coef:.2f}x_{{{j + 1}}}"
    for j, coef in enumerate(row) if coef != 0])
    equation += f" &= {arreglo_13[i]:.2f} \\\\"
    latex_str += equation
latex_str += r"\end{align*}"

display(Latex(latex_str))

# Resolver el sistema de ecuaciones lineales
solucion = np.linalg.solve(arreglo_12, arreglo_13)

# Mostrar el resultado
print(Fore.CYAN + Style.BRIGHT + 
      "Solución del sistema de ecuaciones:" + Style.RESET_ALL)
print(solucion.round(2), "\n")

# Generar LaTeX para mostrar la solución del sistema de ecuaciones
latex_solution = r"\begin{align*}"

for i, value in enumerate(solucion):
    latex_solution += f"x_{{{i+1}}} &= {value:.2f} \\\\"
latex_solution += r"\end{align*}"

display(Latex(latex_solution))
```
