## Ejercicio 06

Seleccione una columna del DataFrame **original** que contenga una variable categórica (si no hay variables categóricas cree una columna con variables categóricas a partir de alguna columna numérica del DataFrame).

Utilice el método **groupby** para analizar la información del DataFrame a partir de la variable categórica seleccionada.

Muestre el resultado y las características del DataFrame.

```python
# Agrupar el DataFrame por la columna 'location' contando
# el número de universidades en cada ubicación
conteo_universidades = df.groupby('location').size()

# Convertir la Serie en un DataFrame
resultados = conteo_universidades.to_frame()

# Cambiar el nombre de la columna
resultados.columns = ['Número de universidades']

# Agrupar el DataFrame por la columna 'location'
# y calcular la media de scores_research
promedio_scores_research = df.groupby('location')['scores_research'].mean()

# Añadir la serie al DataFrame
resultados['Promedio de scores_research'] = promedio_scores_research

# Convertir la columna 'stats_number_students' en string
df['stats_number_students'] = df['stats_number_students'].astype(str)

# Reemplazar ',' por '' en la columna 'stats_number_students'
df['stats_number_students'] = df['stats_number_students'].str.replace(',', '')

# Convertir la columna stats_number_students en numérica
df['stats_number_students'] = pd.to_numeric(df['stats_number_students'], 
                                           errors='coerce')

# Agrupar el DataFrame por la columna 'location'
# y calcular la media de stats_number_students
promedio_estudiantes = df.groupby('location')['stats_number_students'].mean()

# Añadir la serie al DataFrame
resultados['Promedio de estudiantes'] = promedio_estudiantes

# Convertir el 'location' en el indice
resultados.index.name = 'Ubicación'

# Mostrar el DataFrame resultante
print("\033[3m" + "\033[1m" + Fore.RED + 
      "DataFrame agrupado por la columna 'location':" + Style.RESET_ALL)
print(resultados)

# Mostrar las características del DataFrame
print("\n" "\033[3m" + "\033[1m" + Fore.RED + 
      "Características del DataFrame:" + Style.RESET_ALL)
resultados.info(verbose=True)
```

## Ejercicio 07

Cree una tabla dinámica, utilizando el método **pivottable**, para analizar la información del DataFrame **original**.

Muestre el resultado y las características del DataFrame.

```python
# Crear una tabla dinámica con el método pivottable
tabla_dinamica = pd.pivot_table(df, values=['scores_citations', 
                                            'scores_research', 
                                            'stats_number_students'], 
                               index=['location'], 
                               aggfunc={'scores_citations': 'mean', 
                                        'scores_research': 'mean', 
                                        'stats_number_students': 'mean'})

# Mostrar el DataFrame resultante
print("\033[3m" + "\033[1m" + Fore.RED + 
      "Tabla dinámica creada con el método pivottable:" + Style.RESET_ALL)
print(tabla_dinamica)

# Mostrar las características del DataFrame
print("\n" "\033[3m" + "\033[1m" + Fore.RED + 
      "Características del DataFrame:" + Style.RESET_ALL)
tabla_dinamica.info(verbose=True)
```

## Ejercicio 08

Encuentre todos los valores nulos del DataFrame original y cambie esos valores nulos con la media de la respectiva columna.

Muestre el resultado y las características del DataFrame.

```python
# Encontrar los valores nulos del DataFrame original
valores_nulos = df.isnull().sum()

# Mostrar los valores nulos
print("\033[3m" + "\033[1m" + Fore.RED + 
      "Valores nulos del DataFrame original:" + Style.RESET_ALL)
print(valores_nulos)

# Características de las columnas con valores numéricos que contienen NaN
mask = df.columns.str.contains('scores|stats') \
      & ~df.columns.str.contains('rank|male')

# Seleccionar columnas que contengan las características especificadas
cols = df.columns[mask]

# Realizar copia del DataFrame
df_not_nulls = df.copy()

# Dividir los valores de la columna 'scores_overall' en dos nuevas columnas
# Usar 'str.split' y expandimos en dos columnas; 
# si no hay rango, el número se duplicará en ambas columnas
df_not_nulls[['scores_overall_min', 'scores_overall_max']] = \
      df_not_nulls['scores_overall'].str.split('–', expand=True)

# Si no hay guion, 'scores_overall_max' será NaN; 
# se copia 'scores_overall_min' en esos casos
df_not_nulls['scores_overall_max'] = \
      df_not_nulls['scores_overall_max'].fillna(
                                          df_not_nulls['scores_overall_min'])

# Reemplazar '%' por '' en la columna 'stats_pc_intl_students'
df_not_nulls['stats_pc_intl_students'] = \
      df_not_nulls['stats_pc_intl_students'].replace('%', '', regex=True)

# Reemplazar ',' por '' en la columna 'stats_number_students'
df_not_nulls['stats_number_students'] = \
      df_not_nulls['stats_number_students'].replace(',', '', regex=True)

# Convertir las columnas seleccionadas a numérico
df_not_nulls[cols] = df_not_nulls[cols].apply(pd.to_numeric, errors='coerce')

# Convertir las columnas min-max a numérico
col_minmax = ['scores_overall_min', 'scores_overall_max']
df_not_nulls[col_minmax] = df_not_nulls[col_minmax].astype(float)

# Calcular la media del rango para cada fila min-max
df_not_nulls['scores_overall_mean'] = \
      (df_not_nulls['scores_overall_min'] + \
            df_not_nulls['scores_overall_max']) / 2

# Reemplazar 'scores_overall' por 'scores_overall_mean'
# y eliminar las columnas auxiliares
df_not_nulls['scores_overall'] = df_not_nulls['scores_overall_mean']
df_not_nulls.drop(columns=['scores_overall_min', 'scores_overall_max',
                              'scores_overall_mean'], inplace=True)

# Calcular la media y reemplazar los valores nulos con la media de cada columna
df_not_nulls[cols] = df_not_nulls[cols].fillna(df_not_nulls[cols].mean())

# Mostrar el DataFrame resultante
print("\n" + "\033[3m" + "\033[1m" + Fore.RED + 
      "DataFrame con los valores nulos reemplazados por la media:" + 
      Style.RESET_ALL)
print(df_not_nulls)

# Mostrar las características del DataFrame
print("\n" "\033[3m" + "\033[1m" + Fore.RED + 
      "Características del DataFrame:" + Style.RESET_ALL)
df_not_nulls.info(verbose=False)
```
