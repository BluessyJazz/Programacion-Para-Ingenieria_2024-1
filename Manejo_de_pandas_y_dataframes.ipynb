##Estructura de datos: DataFrame

```
import pandas as pd
compra_1 = pd.Series({'Nombre': 'Chris',
                        'Producto Vendido': 'Comida de perro',
                        'Precio': 22.50})
compra_2 = pd.Series({'Nombre': 'Kevyn',
                        'Producto Vendido': 'Comida de gatos',
                        'Precio': 2.50})
compra_3 = pd.Series({'Nombre': 'Vinod',
                        'Producto Vendido': 'Comida de pajaros',
                        'Precio': 5.00})
df = pd.DataFrame([compra_1, compra_2, compra_3], index=['Tienda 1', 'Tienda 1', 'Tienda 2'])
df.head()
```

```
def sumar(v):
    v = v + 1
    return float(v)
df['Precio'] = df['Precio'].map(sumar)
df
```

```
df.columns
```

```
for col in df.columns:
    if col[:2]=='01':
        df.rename(columns={col:'Oro' + col[4:]}, inplace=True)
    if col[:2]=='02':
        df.rename(columns={col:'Plata' + col[4:]}, inplace=True)
    if col[:2]=='03':
        df.rename(columns={col:'Bronce' + col[4:]}, inplace=True)
    if col[:1]=='№':
        df.rename(columns={col:'#' + col[1:]}, inplace=True)

df.head()
```
